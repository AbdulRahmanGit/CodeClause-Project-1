  Separate formatted data into Year, Month  

```python
#In this code, we have sorted our data by date
df['Date']=pd.to_datetime(df['Date'], utc=True)
df['date'] = [d.date() for d in df['Date']]
df
```

```python
correct_date=pd.to_datetime(df['date'], format='%Y-%m-%d').dt.strftime('%m-%d-%Y')
df1=pd.DataFrame(correct_date)
df['date']=df1['date'].astype('datetime64[ns]')
df2=df['date'].dt.year
df['year']=df2.astype('string')
df
```

In this line of code, we separate the date and assign  separate column named date .
    

```python
df3=df['date'].dt.month
month=[]
for i in df3:
    if i ==1:
        month.append("January")
    if i ==2:
        month.append("February")
    if i ==3:
        month.append("March")
    if i ==4:
        month.append("April")
    if i ==5:
        month.append("May")
    if i ==6:
        month.append("June")
    if i ==7:
        month.append("July")
    if i ==8:
        month.append("August")
    if i ==9:
        month.append("September")
    if i ==10:
        month.append("October")
    if i ==11:
        month.append("November")
    if i ==12:
        month.append("December")


df3=pd.DataFrame(month)
df['month'] = df3
df
```

 In this part of the code, we separate the months and put a column called month

```python
Df = df[['Index','date','year',"month","Open","High","Low", 'Close','Adj Close','Volume']]
Df
```
